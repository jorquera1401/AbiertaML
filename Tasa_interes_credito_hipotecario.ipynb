```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import csv

data = []

with open('data.csv', newline = '') as csvFile:
  csv_reader = csv.reader(csvFile, delimiter='\n')

  for row in csv_reader:
    data.append(row)

print(data)
print(len(data))


```

```python
import random

month = 6

X_train = []
y_train = []
X_test = []

for i in range(len(data) - month - 1):
  sequence = data[i:(i+month)]
  target = data[i+month]

  flat = [value[0] for value in sequence]

  X_train.append(flat)
  y_train.append(target)


for lista in X_train:
    test = []
    for valor in lista:
      random_number = random.uniform(-1, 1)
      element = float(valor) + random_number
      test.append(element)
    X_test.append(test)

print("X_test", X_test)
print("X_train", X_train)


```

```python
model = LogisticRegression()
model = LogisticRegression(max_iter=100000)
model.fit(X_train, y_train)


y_test = y_train


y_pred = model.predict(X_test)

# Calcula la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo:", accuracy)

last = data[-month:]
new_last = []

for lista in last:
  for valor in lista:
    new_last.append(float(valor))

print("tasa de interes en los ultimos", month, " meses : ",new_last)


y_future_pred =model.predict([new_last])
print('future: en ', month, 'meses - valor:', y_future_pred)
```
